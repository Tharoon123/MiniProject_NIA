# Handling Missing Data

```python
df.info()
```

```python
df['Rate']=pd.to_numeric(df['Rate'],errors='coerce')
df.info()
```

```python
df['Rate'].mean()
df[['Rate','Duration']].corr()
```

```python
df.isnull().sum()
```

```python
m_Rate=df['Rate'].mean()
df['Rate']=df['Rate'].fillna(m_Rate)
df.isnull().sum()
#df['Rate']
```

```python
m_Duration=df['Duration'].mean()
df['Duration']=df['Duration'].fillna(m_Duration)
df.isnull().sum()
#df[['Rate','Duration']]
```

```python
#df['Nudity'].unique()

m_Nudity=df['Nudity'].mode()[0]
df['Nudity']=df['Nudity'].fillna(m_Nudity)
df.isnull().sum()
```

```python
m_Certificate=df['Certificate'].mode()[0]
df['Certificate']=df['Certificate'].fillna(m_Certificate)
df.isnull().sum()
```

```python
#df['Violence']
#df['Violence'].unique()

m_Violence=df['Violence'].mode()[0]
#m_Violence
df['Violence']=df['Violence'].fillna(m_Violence)
df.isnull().sum()
```

```python
#df['Profanity']
#df['Profanity'].unique()

m_Profanity=df['Profanity'].mode()[0]
#m_Profanity
df['Profanity']=df['Profanity'].fillna(m_Profanity)
df.isnull().sum()
```

```python
#df['Alcohol']
#df['Alcohol'].unique()

m_Alcohol=df['Alcohol'].mode()[0]
#m_Alcohol
df['Alcohol']=df['Alcohol'].fillna(m_Alcohol)
df.isnull().sum()
```

```python
#df['Frightening']
#df['Frightening'].unique()

m_Frightening=df['Frightening'].mode()[0]
#m_Frightening
df['Frightening']=df['Frightening'].fillna(m_Frightening)
df.isnull().sum()
df.head(10)
```

```python
print(df.info(),df['Votes'].head(10))
```

```python
df['Votes']=pd.to_numeric(df['Votes'].replace({'No Votes':'',',':''},regex=True) ,errors='coerce')
df['Votes'].head(10)
#df.isnull().sum()
vote_mean=df['Votes'].mean()
df['Votes'].fillna(vote_mean,inplace=True)
df.isnull().sum()
df['Votes'].head(10)
```

```python
from sklearn.cluster import KMeans
k_values = [1,2,3,4,5]
WCSS =[]
#to check the k_values one by one
for k in k_values:
  model = KMeans(n_clusters=k)
  model.fit(dfnew[['Rate','Nudity_1','Violence_1','Profanity_1','Alcohol_1','Frightening_1']])
  WCSS.append (model.inertia_)
```

```python
model = KMeans(n_clusters=3)
pred = model.fit_predict(dfnew[['Violence_1','Frightening_1']])
pred
```

```python
dfnew['clusters'] = pred
dfnew.head(5)
```

```python
c1=dfnew[dfnew['clusters']==0]
c2=dfnew[dfnew['clusters']==1]
c3=dfnew[dfnew['clusters']==2]
```

# Linear Regression

```python
dfImputedData=dfnew.drop(['Name'],axis=1)
dfImputedData.head(10)
```

```python
corr_matrix = dfImputedData.corr()
corr_with_target = corr_matrix['Rate'].sort_values(ascending=False)
print(corr_with_target)
```

```python
plt.figure(figsize=(10, 6))
corr_with_target.drop('Rate').plot(kind='bar', color='skyblue')
plt.title('Correlation with Target Variable (Rate)')
plt.xlabel('Features')
plt.ylabel('Correlation')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
dfImputedData.head(10)
```

```python
features = dfImputedData.columns.drop(['Nudity_1','Alcohol_1','Profanity_1','clusters','Rate']).tolist()

y = dfImputedData['Rate']
X = dfImputedData[features]
features
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

scaler = MinMaxScaler().set_output(transform="pandas")
X_normalized = scaler.fit_transform(X)

X_normalized.columns
```

```python
from sklearn.model_selection import train_test_split
train_X, val_X, train_y, val_y = train_test_split(X_normalized, y, random_state = 10 ,test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(train_X, train_y)

```

```python
from sklearn.metrics import mean_absolute_error

val_predictions = model.predict(val_X)
print(mean_absolute_error(val_y, val_predictions))
```
